# Análise da curva de aprendizado para 50x50

```python
import pandas as pd
df = pd.read_csv('results/statistics_50_50.csv')
df.head()
```

```python
import seaborn as sns
sns.lineplot(data=df, x="episode", y="rewards").set(title="Reward por episódios")
```

```python
sns.lineplot(data=df, x="episode", y=df['rewards'].ewm(span=100).mean()).set(title="Reward por episódios (valor médio considerando janela de 100 episódios)")
```
